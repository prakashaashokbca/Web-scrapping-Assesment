```python
import requests
```

```python
from bs4 import BeautifulSoup    
```

```python
import pandas as ap
```

## Scrapping Website

```python
url = "https://www.flipkart.com/search?q=camera&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

## Website Response

```python
res= requests.get(url)
```

```python
res
```

## Parsing

```python
soup = BeautifulSoup(res.text)
```

```python
soup
```

##  camera names

```python
camera = soup.select("._4rR01T")
```

```python
camera
```

```python
prices = soup.select('._1_WHN1')
```

```python
prices
```

```python
price_title=[]
```

```python
for item in prices:
    price_title.append(item.text)
```

```python
price_title
```

```python
camera_names=[]
```

```python
for item in camera:
    camera_names.append(item.text)
```

```python
camera_names
```

## Data Frames

```python
df=ap.DataFrame(camera_names,columns=['camera names'])
```

```python
df
```

```python
df['price_title']=price_title
```

```python
df
```

```python
df.to_csv('pr.csv')
```
